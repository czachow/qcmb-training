## Welche Arten von Zahlen kennen Sie?

Natürlich die

### Natürliche Zahlen:  $ \mathbb{N} $

$$
0, 1, 2, 3, … , 9302, 9303, …. 
$$

Diese sind „abgeschlossen“ bezüglich Plus / Addition, allerdings nicht bezüglich Minus / Subtraktion. 
Dafür gibt es:

### Ganze Zahlen: $ \mathbb{Z} $

$$
... , -9303, -9302, …. -3, -2, -1, 0, +1, +2, +3, … , +9302, +9303, …. 
$$

Die ganzen Zahlen sind sogar abgeschlossen bezüglich Malnehmen / Multiplikation, 
allerdings nicht bezüglich Teilen / Dividieren. Dafür gibt es 

### Brüche, rationale Zahlen (Ganze Zahl / Ganze Zahl): $ \mathbb{Q} $

$$
..., -3, -\frac{19}{9},-2, -\frac{17}{9}, -1, -\frac{3}{4}, -\frac{1}{4}, 0, +\frac{1}{4}, +\frac{3}{4}, +1, +\frac{17}{9}, +2, +\frac{19}{9}, +3, ...
$$

(das sind im Prinzip auch alle „normalen“ Kommazahlen: 1.25, 76.3456772123, 1.3333333333….)

Sind das alle Zahlen, mit denen wir es normalerweise zu tun haben? Also alle Zahlen, die man an einem Meterband ablesen kann? Nein! Könnten wir alle Brüche als winzige Punkte an einem Meterband einzeichnen, würde wir danach viele Lücken sehen. Es gibt sogar sehr viel mehr Lücken als Punkte! Die Brüche sind nicht abgeschlossen bezüglich ihrer „Grenzwerte“ (die Lücken): Es fehlen noch die „irrationalen“ Zahlen $ \sqrt{2}, \pi, 2*\pi, 3*\pi, e, 2*e, 3*e, ...$ (mehr zur Zahl "e" weiter unten).

Zusammen sind das 

### Reelle Zahlen, alle „normalen“ Zahlen:  $ \mathbb{R} $

<br/>
 <center> Alle Brüche + ihre Lücken (die „irrationalen“ Zahlen) </center>
<br/>

Die reellen Zahlen, sind im Prinzip alle „normalen“ Zahlen, die uns im Alltag überhaupt begegnen können. Aber sind das auch alle Zahlen, die man benötigt, um erfolgreich Mathematik zu betreiben? Nein! Die reellen Zahlen sind nicht abgeschlossen gegenüber „Wurzelbildung“: Z.b. 


**Finde die Zahl „$i$“ mit **

$$
i*i = i² = -1
$$

<center> **bzw.**  </center> 

$$
i = \sqrt{-1}
$$


Bereits im frühen Mittelalter erkannten damalige Mathematiker: 
Wenn man hierfür eine *neue* Zahl **„$i$“ für „imaginär“ ** als Hilfe einfach definiert, kann man plötzlich gewisse Aufgaben lösen, deren Lösung dann wieder eine „normale“, relle Zahl ist.
Kombiniert man die Zahl „$i$“ mit den reellen Zahlen erhält man einen neuen erweiterten Zahlenbereich: 



## Komplexe Zahlen: $ \mathbb{C} $

$$
z  =  x  + i * y 
$$

<center> (eine **"komplexe Zahl" $z$**, <br/>
wobei der „Realteil“ $x$ und der „Imaginärteil“ $y$ normale, reelle Zahlen sind) </center>

Nebenbei ist jede reelle Zahl auch eine komplexe Zahl aber ohne Imaginäranteil.
Im Laufe der Zeit fand man für die komplexen Zahlen so viele erstaunliche und wichtige Eigenschaften heraus, dass sie bereits Ende des 18. Jahrhunderts ein fester, wichtiger Bestandteil der Mathematik wurden.
Selbst in der Physik werden komplexe Zahlen verwendet, um viele Berechnungen zu vereinfachen: 
Z.b. um Wellen, zu beschreiben. Und für die Quantenphysik hat sich herausgestellt: 
Die komplexen Zahlen sind zur Beschreibung der Quantenwelt unerläßlich! 
Der Grund hierfür ist die sogenannte „Unschärferelation“, die Werner Heisenberg 1927 herausfand. 
Im Quantencomputing sind die komplexen Zahlen eine wichtige Ursache für den exponentiellen Rechenvorteil gegenüber herkömmlichen Computern bei gewissen Aufgaben.

### Rechenregeln für komplexe Zahlen

Folgende Regeln gibt es dabei für zwei komplexe Zahlen $z_1$ und $z_1  \epsilon  \mathbb{C}$

$$
z_1  =  x_1  + i * y_1 \\
z_2  =  x_2  + i * y_2 
$$

#### Addition (gilt auch für Minus): 

$$
z_1 + z_2  =   (x_1  + x_2)  +  i * (y_1  + y_2) 
$$

<center>(Real- und Imginärteil addieren sich also jeweils)</center>

#### Multiplikation:

$$
z_1 * z_2  =  (x_1  +  i * y_1) * (x_2  +  i * y_2)
$$

<center>einfach ausmultiplizieren und $i * i = -1$ verwenden:</center>

$$
=  (x_1*x_2 - y_1*y_2)  +  i * (x_1*y_2 + y_1*x_2) 
$$

#### „konjugiert“ komplexe Zahl  $z^*$

Die „Konjugation“ einer komplexen Zahl $z^*$ (manchmal auch $\bar{z}$) ist definiert durch 

$$
z^*  = x -  i * y
$$

<center>(beachten Sie das Minus-Zeichen)</center>

#### Länge bzw. den „Betrag“ |z|
Damit berechnet man die Länge bzw. den „Betrag“ $|z|$ einer komplexen Zahl:

$$
|z|^2 = z * z^* =   (x  +  i * y) * (x  -  i * y) = x^2 + y^2 +  i * 0 
$$

## Die Komplexe Zahlenebene und die Polardarstellung


Die imaginäre Einheit „$i$“ findet man natürlich nicht auf einem normalen Metermaß. 
Dafür führt man einen zusätzlichen „Freiheitsgrad“ ein: 
Ein zweites Metermaß, das zusammen mit den rellen Zahlen eine Ebene aufspannt: 
Die **„Komplexe Zahlenebene“**. Die komplexe Zahl ist darin ein Zeiger:

*Eine komplexe Zahl als Zeiger in der „Komplexen Zahlenebene“*

Jetzt müssen wir uns noch an ein paar Eigenschaften über die Seitenlängen von rechtwinkligen Dreiecken erinnern. 
Nach dem **Satz des Pythagoras** gilt z.B. für die **Länge bzw. der Betrag |z|** dieses Zeigers:

$$
|z|^2 = x^2 + y^2
$$

<center> bzw. </center>

$$
|z| = \sqrt{x^2 + y^2}
$$

Außerdem kann man dem Zeiger einen Winkel φ zuordnen ($\varphi$ = „Phi“: Achtung Physiker lieben griechische Buchstaben!).
Für rechtwinklige Dreiecke gibt es auch die Sinus- und Kosinusfunktionen: Die **"trignometrischen Funktionen"** . Diese sind gerade so definiert, dass man die fehlenden Dreiecksseiten aus der Grundseite und dem Winkel berechnen kann:

$$
Re(z) = x = |z| * cos(\varphi) \\
Im(z) = y = |z| * sin(\varphi)
$$

Nebenbei: Setzt man dies in den Satz des Pythagoras ein, erhält man die wichtige Eigenschaft:

$$
sin^2(\varphi) + cos^2(\varphi) = 1
$$

Jede **komplexe Zahl $z$** lässt sich also auch durch die Zeigerlänge $|z|$ und den Winkel in der komplexen Ebene darstellen. Dies nennt man auch die „Polardarstellung“ der komplexen Zahl:

\begin{align*}
  z &= x  + i * y\\
    &= |z| cos(\varphi) + i * |z| sin(\varphi) 
\end{align*}

**<center>(„Polardarstellung“ einer komplexen Zahl)</center>**


## Die Eulersche Formel und die Eulersche Exponentialfunktion

Im Jahr 1748 veröffentlichte der schweizer Mathematiker Leonhard Euler folgende Gleichung für komplexe Zahlen, die u.a. auch für die Quantenmechanik von wichtiger Bedeutung ist:

$$
 exp(i*\varphi) = e^{i*\varphi} = cos(\varphi) + i*sin(\varphi)
$$

<center>**(„Eulersche Formel“)**</center>

„$exp(x)$“ bzw. „$e^x$“ ist die sogenannte **„Eulersche Exponentialfunktion“**. 
Und auf der rechten Seite der Formel erkennen Sie die Polardarstellung einer komplexen Zahl *mit Länge 1*.
Um es vorweg zu nehmen: Bildlich sagt die Eulersche Formel Folgendes aus (das werden Sie nach dem folgenden Exkursen über Exponentialfunktionen erkennen):

### Exkurs: Was sind „Exponentialfunktionen“?

„Exponentialfunktionen“ $a^{x}$ (für reelle Zahlen a und x)  spielen eine wichtige Rolle in der Mathematik und in den Naturwissenschaften 
(z.B.  für $a=10: 10^{x}$ oder für $a=2: 2^{x}$). Sie beschreiben normalerweise ein exponentielles Wachstum. Folgende Regeln gelten dabei:

$$
a^{x+y} = a^{x} * a^{y}
$$

<center> z.B. </center>
$$
100 000 = 10^{5} = 10^{3} * 10^{2} = 1000 * 100
$$

Eine Exponentialfunktion macht also aus der Addition von zwei *reellen* Zahlen eine Multiplikation. 
Außerdem gilt:

$$
 (a^{x})^{y}= a^{x*y}
$$


<center> z.B. </center>
$$
1 000 000 = 10^{6} = 10^{3*2} = (10^{3})^2 = 1000^2 = 1000 * 1000
$$

<center> oder auch </center>
$$
\sqrt{1 000 000} = (10^{6})^{1/2} = 10^{6*1/2} = 10^{3} = 1000
$$

Eine Exponentialfunktion macht also aus der Multiplikation von *zwei reellen* Zahlen eine „Exponentiation“.

### Exkurs: Was ist die „Eulersche Exponentialfunktionen“?

Die „Eulersche Zahl e“ aus der Formel oben ist eine ganz spezielle Zahl: 
$e = 2.7182...$, deren Wert für uns eigentlich gar nicht wichtig ist. 
Wichtig für uns ist nur, was die Eulersche Formel aussagt. Die Exponentialfunktion der Zahl „$e$“ hat nämlich folgende die wichtige Eigenschaft: 

Verwendet man darin anstelle von einer reellen Zahl $x$, also $e^{x}$, einen *ein imaginären „Winkel“* $i*\varphi$, also $e^{i*\varphi}$, so **dreht** die Eulersche Exponentialfunktion jede komplexe Zahl $z$ durch einfache Multiplkation $z * e^{i*\varphi}$ gerade um den Winkel $\varphi$ in der komplexen Zahlenebene!
Anstelle von einem exponentiellen Wachstum **führt die Eulerfunktion für imaginäre Zahlen also Drehungen aus**.
*Übrigens:* Der Trick mit den Drehungen und der Eulersche Exponentialfunktion wird Ihnen später wieder begegnen. 
So werden nämlich beliebige Quantenzustände in neue Quantenzustände „gedreht“. 
In dem Fall aber nicht über einen einfachen imaginären Winkel, sondern durch eine „Matrix“. 
Dazu aber später mehr.

## Der Winkel π (Pi) und die „schönste Gleichung der Mathematik“

In der Mathematik und in der Physik ist es üblich Winkel in Einheiten von der "Kreiszahl" $\pi=3.14159...$ zu schreiben.


  * $1/2 \pi$ = eine Vierteldrehung
  * $\pi$ = eine halbe Drehung
  * $2 \pi$ = eine ganze Drehung
  * $4 \pi$ = zwei ganze Drehungen
  
Die letzten beiden Drehungen eines Zeigers sehen in der komplexen Ebene vollkommen gleich aus.

Die Identität

$$
e^{i*\pi} = -1
$$

wird übrigens oft als die „schönste Gleichung der Mathematik“ bezeichnet. 
Sie vereint gleich vier grundlegende Größen der Mathematik in einer sehr kompakten Form miteinander.

Abschließend sollten wir noch folgende wichtige Eigenschaft für die Konjugation einer komplexen Zahl erwähnen:

$$
z = |z| * e^{i*\varphi} \\
z^* = |z| * e^{-i*\varphi}
$$

Beachten Sie das Minuszeichen. Dies erkennt man durch:

$$
|z|^2 = z * z^*
$$

<center>und vergleicht das mit</center>

$$
{|z|}^2 = |z| * |z| * e^{i*0} = |z| * |z| * e^{i*\varphi -i*\varphi} = |z| * e^{i*\varphi} * |z| * e^{-i*\varphi}
$$

Darüber erkennen Sie auch, dass der Betrag der Euler-Funktion für rein imaginäre Winkel immer 1 ist:

$$
| e^{i*\varphi} |^2 = e^{i*\varphi} * e^{-i*\varphi} = e^{i*0} = 1
$$

Andersherum gibt es für jede komplexe Zahl $z$ *mit Betrag 1* einen geeigneten Winkel $\varphi$ mit

$$
 |z| = 1 \Rightarrow  z = e^{i*\varphi}
$$
